```python
from gliner import GLiNER
import torch

# Charger le modèle (le modèle multi-langue est recommandé pour le français)
model = GLiNER.from_pretrained("numind/gliner_multi-v2.1")
model.to("cuda" if torch.cuda.is_available() else "cpu")

# Vos labels spécifiques
labels = ["missions du poste", "avantages sociaux", "outils informatiques"]

# Exemple d'application sur une description d'offre
text = "Nous proposons un poste de Data Scientist. Vos missions : coder en Python et SQL. Avantages : Télétravail 3 jours/semaine et mutuelle gratuite."

entities = model.predict_entities(text, labels, threshold=0.5)

for entity in entities:
    print(f"{entity['label']} => {entity['text']}")
```

```python
from gliner import GLiNER
import torch

# Utilisation du repo 'urchade' au lieu de 'numind'
model = GLiNER.from_pretrained("urchade/gliner_multi-v2.1")

# Si vous avez un GPU local (Nvidia), activez-le :
if torch.cuda.is_available():
    model = model.to("cuda")
```

```python
# 1. Vérification du chargement (cas où vous relancez la cellule)
device = "cuda" if torch.cuda.is_available() else "cpu"
# On assume que 'model' est déjà chargé comme dans votre message précédent

# 2. Définition d'un exemple type de votre base JOCas
job_description = """
En tant que Consultant Data, vos missions principales seront :
- Analyser les besoins métiers et rédiger les spécifications techniques.
- Concevoir et développer des pipelines ETL sous Python et SQL.
- Créer des tableaux de bord interactifs avec Power BI.

Le profil recherché possède une expérience de 3 ans. 
Nous offrons un package attractif : télétravail partiel (2 jours), 
tickets restaurant, mutuelle prise en charge à 100% et un accès illimité à une salle de sport.
"""

# 3. Choix des labels (vous pouvez les modifier selon vos besoins de recherche)
# GLiNER comprend très bien les descriptions naturelles
labels = [
    "tâches et missions", 
    "outils et logiciels", 
    "télétravail", # Label dédié
    "avantages sociaux (tickets restaurant, mutuelle, rtt)", # Aide sémantique
    "années d'expérience"
]

# 4. Exécution de la prédiction
# Le seuil (threshold) à 0.4 est un bon point de départ pour éviter les oublis
entities = model.predict_entities(job_description, labels, threshold=0.4)

# 5. Affichage propre des résultats
print(f"--- Analyse de l'offre ({device}) ---\n")
for entity in entities:
    print(f"[{entity['label'].upper()}] : {entity['text']}")

# 6. Test de regroupement par catégorie (pour votre future base de données)
from collections import defaultdict
results_structured = defaultdict(list)
for ent in entities:
    results_structured[ent['label']].append(ent['text'])

print("\n--- Données structurées ---")
import json
print(json.dumps(results_structured, indent=2, ensure_ascii=False))
```
